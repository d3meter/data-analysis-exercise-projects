```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stats
import numpy as np
import math as m
```

# RE California Database

```python
data = pd.read_csv("data-sets/california-properties-dataset.csv", sep=";")
data.columns = [col.strip() for col in data.columns]
data.head(10)
```

```python
# Create a copy and format columns
convert_data = data.copy()
convert_data["Price"] = convert_data["Price"].str.replace("$", "", regex=False).str.replace(",", ".", regex=True)
convert_data["Price"] = convert_data["Price"].apply(lambda x: float("".join(x.split())))

convert_data["Status"] = convert_data["Status"].str.strip()
convert_data["Status"].fillna("Not Sold", inplace=True)

convert_data["Area (ft.)"] = convert_data["Area (ft.)"].str.replace(",", ".").astype(float)

def convert_float_to_int(column_names, dataframe):
    for column_name in column_names:
        dataframe[column_name] = dataframe[column_name].fillna(0).astype(int)
    return dataframe

columns_convert = ["Year of sale", "Month of sale", "Y", "M", "D", "Deal satisfaction"]
convert_data = convert_float_to_int(columns_convert, convert_data)

convert_data
```

## Histograms. Graphing numerical data

<span style="color:gray">The histograms point to similar insights - most of the properties' prices are concentrated between 200,000 and 300,000 dollars.<span/>

## Scatter plot. Plotting numerical data; establishing relationship between variables
- Create a scatter plot showing the relationship between Price and Area. Use the data on all properties, no matter if sold or not. Interpret the results.

```python
fig, ax = plt.subplots(figsize=(8, 4))

x = convert_data["Area (ft.)"]
y = convert_data["Price"]

ax.scatter(x, y, color=primary_color)
ax.set_xlabel("Area in ft.")
ax.set_ylabel("Price in $")
ax.set_title("Relationship Between Property Price and Area")

coefficients = np.polyfit(x, y, 1)
trend_line = np.poly1d(coefficients)

ax.plot(x, trend_line(x), linestyle="-", color=secondary_color, lw=5, alpha=0.6)

plt.show();
```

<span style="color:gray">The plot shows a very strong linear relationship between price and area. At lower prices, the points are closer together. This means that the variance is smaller than at higher prices, so the bigger the property, the bigger the difference in the price.<span/>

## Measures of relationship between variables
- Calculate the covariance and correlation coefficient between Price and Area, no matter if the property is sold or not. Is the result in line with the scatter plot?

```python
covariance = convert_data["Price"].cov(convert_data["Area (ft.)"])
correlation_coefficient = convert_data["Price"].corr(convert_data["Area (ft.)"])

print(f"As the correlation coefficient is very high ({round(correlation_coefficient, 2)}), it represents the same effect as the scatter plot above: the linearity between Price and Area is very strong. The covariance is {round(covariance)}")
```

## Frequency distribution table. Pareto diagram.
- Create a frequency distribution table, where you list all the countries from which the company has buyers (country vs number of buyers). Count the absolute frequency, the relative frequency and the cumulative frequency.
- Create a Pareto diagram representing the data.

```python
def create_frequency_table(column_name, dataframe):
    keys = dataframe.loc[convert_data[column_name].notna(), column_name].unique()
    frequency = []
    curr_freq = 0
    sum_freq = 0
    
    for key in keys:
        curr_freq = len(convert_data.loc[dataframe[column_name] == key])
        frequency.append(curr_freq)
        sum_freq += curr_freq
        
    frequency_table = pd.DataFrame({"Frequency": frequency}, index=keys)
    frequency_table = frequency_table.sort_index()
    frequency_table.sort_values(by="Frequency", ascending=False, inplace=True)
    
    # Calculate and add relative and cumulative frequencies
    frequency_table["Relative Frequency [%]"] = round(frequency_table["Frequency"] / sum_freq * 100, 2)
    frequency_table["Cumulative Frequency [%]"] = frequency_table["Relative Frequency [%]"].cumsum() 
    
    return frequency_table

freq_distribution_table = create_frequency_table("Country", convert_data)
freq_distribution_table
```

```python
fig, ax1 = plt.subplots(figsize=(8, 6))

x = freq_distribution_table.index
y = freq_distribution_table["Frequency"]
y_sec = freq_distribution_table["Cumulative Frequency [%]"]

ax1.bar(x, y, color=secondary_color)

ax1.set_title("Market share")
ax1.set_xlabel("Countries")
ax1.set_ylabel("Frequency")

for i, v in enumerate(y):
    ax1.text(i, v + 1, str(v), ha='center', va='bottom')

# Add pareto line
ax2 = ax1.twinx()
ax2.plot(x, y_sec, color=primary_color, lw = 3, marker="o")

ax2.set_yticks(y_sec)
ax2.set_yticklabels([f"{round(val, 1)}%" for val in y_sec])

plt.show();
```

## Measures of central tendency, asymmetry and variability
- Calculate the mean, median, mode, skewness, variance and standard deviation of Price for all properties, no matter if sold or not.
- Interpret the measures.

```python
properties_price_mean = round(stats.mean(convert_data["Price"]), 2)
properties_price_median = round(stats.median(convert_data["Price"]), 2)
properties_price_mode = round(stats.mode(convert_data["Price"]), 2)
properties_price_stddev = round(stats.stdev(convert_data["Price"]), 2)

properties_price_skewness = round(convert_data["Price"].skew(), 2)
properties_price_variance = round(stats.variance(convert_data["Price"]), 2)

analysis_indexes = ["Mean", "Median", "Mode", "Standard Deviation", "Skewness", "Variance"]
analysis_values_with_currency = [properties_price_mean, properties_price_median, properties_price_mode, properties_price_stddev]
analysis_values_without_currency = [properties_price_skewness, properties_price_variance]

format_currency = lambda x: f"$ {x:,.2f}".replace(',', ' ')
analysis_values_with_currency = [format_currency(value) for value in analysis_values_with_currency]

format_no_currency = lambda x: f"{x:,.2f}".replace(',', ' ')
analysis_values_without_currency = [format_no_currency(value) for value in analysis_values_without_currency]

analysis_values = analysis_values_with_currency + analysis_values_without_currency

properties_price_analysis = pd.DataFrame({"Properties' Price Analysis": analysis_values}, index=analysis_indexes)


properties_price_analysis
```

```python
plt.figure(figsize=(8, 6))
plt.title("Density of Properties' Prices")
plt.xlabel("Price Range")
plt.ylabel("Density")

convert_data["Price"].plot.density(color=primary_color, linewidth=2)

plt.xlim(left=0)

plt.axvline(x=properties_price_median, color=secondary_color, linestyle="--", label="Median")
plt.axvline(x=properties_price_mean, color="deepskyblue", linestyle="--", label="Mean")
plt.axvline(x=properties_price_mode, color="cadetblue", linestyle="--", label="Mode")

plt.legend()

plt.grid(axis="y")
plt.show();
```

<span style="color:gray">The skewness is right-tailed (positive) because the mean is greater than the Median. In other words, there is a higher proportion of cheaper properties than more expensive properties.The distribution of property prices is influenced by some relatively expensive properties that pull the mean upwards.<span/>
