```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_white
```

```python
#Define colors of plots
primary_color = "lightseagreen"
secondary_color = "skyblue"
```

# Regression Analysis

```python
data = pd.read_csv("data-sets/car-sales-data.csv", sep=";")
data.columns = [col.strip() for col in data.columns]
data.head(10)
```

## Data cleaning

```python
# Create a copy and format columns
convert_data = data.copy()
```

```python
def convert_accounting_to_float(column_names, dataframe):
    for column_name in column_names:
        dataframe[column_name] = dataframe[column_name].str.replace("$", "", regex=False).str.replace(",", ".", regex=True)
        dataframe[column_name] = dataframe[column_name].apply(lambda x: float("".join(x.split())))
    return dataframe

columns_to_float = ["Price"]

convert_data = convert_accounting_to_float(columns_to_float, convert_data)

convert_data.head(10)
```

```python
def convert_string_to_float(column_names, dataframe):
    for column_name in column_names:
        dataframe[column_name] = dataframe[column_name].str.replace(",", ".", regex=True).astype(float)
        
    return dataframe


string_cols_to_float = ["EngineV"]

convert_data = convert_string_to_float(string_cols_to_float, convert_data)

convert_data.head(10)
```

## Check linearity, normality and homoscedasticity

## Preform linear regression analysis

```python
def perform_linear_regression(data, X_columns, Y_column):

    X = data[X_columns]
    X = sm.add_constant(X)
    
    y = data[Y_column]
    
    model = sm.OLS(y, X).fit()
    model_summary = model.summary().as_text()
    
    return model_summary
```

### Price - Mileage result

```python
price_mileage_result = perform_linear_regression(convert_data, "log_mileage", "log_price")

print(price_mileage_result)
```

#### Conclusion
- The model is significant, because Prob (F-statistic) is very close to 0.
- The explanatory variable is signicant, because the P-Value is 0.
- The explanatory power of our model could be stronger, because R-square adj. is only 0.27.

### Price - Engine Volumen result

```python
price_enginev_result = perform_linear_regression(convert_data, "EngineV", "log_price")

print(price_enginev_result)
```

#### Conclusion
- The model is significant, because Prob (F-statistic) is very close to 0.
- The explanatory variable is signicant, because the P-Value is 0.
- The explanatory power of our model could be stronger, because R-square adj. is only 0.21.

### Price - Year result

```python
price_year_result = perform_linear_regression(convert_data, "Year", "log_price")

print(price_year_result)
```

#### Conclusion
- The model is significant, because Prob (F-statistic) is very close to 0.
- The explanatory variable is signicant, because the P-Value is 0.
- The explanatory power is much more stronger, because R-square adj. is 0.56.

### Price - Mileage, Engine Volumen, Year  result

```python
x_cols = ["log_mileage", "EngineV", "Year"]

price_all_result = perform_linear_regression(convert_data, x_cols, "log_price")

print(price_all_result)
```

#### Conclusion

- The explanatory variable is signicant, because the P-Value is 0.

## Add dummy variables

```python
brands = convert_data["Brand"].unique()
brand_col_names = [brand.lower() for brand in brands]

brand_col_names
```

```python
for brand_name in brand_col_names:
    convert_data[brand_name] = convert_data.apply(lambda row: 1 if row['Brand'].lower() == brand_name else 0, axis=1)
        
    
convert_data
```

## Regression analysis incl. brands
Volkswagen is the benchmark against which we are compared. Number of dummies always should be 1 less than the options.

```python
x_cols_brand = ["log_mileage", "EngineV", "Year", "mercedes-benz", "toyota", "bmw", "audi", "mitsubishi", "renault"]

price_all_brand_result = perform_linear_regression(convert_data, x_cols_brand, "log_price")

print(price_all_brand_result)
```

The new model has an increased explenatory power and it shows prices vary according to different parameters. For example: +1 year: +4% price increase or Mercedes-Benz is 10+1 = 11% more expensive but Renault is 16+1 = 17% cheaper than Volkswagen (our benchmark brand).
